## Data processing

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")  # silence warnings
```

```python
data_path = '/data/Data/UKBB/New2/ukb670081.csv'
```

```python
df_chunks = pd.read_csv(data_path, chunksize=2000,low_memory=False,encoding='latin1')
```

```python
temp_chunk = df_chunks.get_chunk()
temp_chunk
```

```python
[string for string in temp_chunk.columns if "21001" in string]
```

```python
ICD10 = [string for string in temp_chunk.columns if "41270" in string]
```

```python
cols = ['eid',
        '21003-2.0', # Age
        '22001-0.0', #Genetic Sex
        '21001-2.0', #BMI
        '25005-2.0', #Volume of gray matter normalized for head size
        '4079-2.0', #Diastolic blood pressure- reading 1
        '4079-2.1', #Diastolic blood pressure- reading 2
        '4080-2.0', #Systolic blood pressure- reading 1
        '4080-2.1', #Systolic blood pressure- reading 2
        '48-2.0', #Waist circumference        
        '49-2.0', #Hip circumference
        '23099-2.0', #BF%
        '2443-2.0', #Diabetes
        '4056-2.0', # Age stroke diagnosed
        '20116-2.0', #Smoking status
        '1558-2.0', # Alcohol intake frequency
        '894-2.0', #Duration of moderate activity
        '914-2.0', #Duration of vigourous activity       
        ] + ICD10
```

```python
fdf = []

for chunk in tqdm(df_chunks):
    filtered_chunk = chunk.filter(items=cols)
    filtered_chunk = filtered_chunk[filtered_chunk[[
       '25005-2.0'
        ]].notnull().all(1)]
    fdf.append(filtered_chunk) 
```

```python
filtered_df = pd.concat(fdf)
filtered_df = filtered_df.reset_index(drop=True)
filtered_df
```

```python
filtered_df.to_csv('Paper/ukbb.csv', index=False)
```
